```python
import pandas as pd
import numpy as np
```

```python
dfs = pd.read_csv('DataSet1_1h_Fev01,22_Mar02,22.csv')
#pd.set_option("display.max_rows", None, "display.max_columns", None)
dfs
```

```python
def sma(data, window):
    return data.rolling(window = window).mean()

def bollinger_bands(data,sma,window,nstd):
    std = data.rolling(window = window).std()
    upper_band = sma + std * nstd
    lower_band = sma - std * nstd
    return upper_band,lower_band
```

```python
symbols = ['BTC','DOGE','ADA','LUNA','SAND','ETH','MATIC','AVAX','SUPER']
nsdt = 3
MA_window = 20

for symbol in symbols:
    dfs[f'{symbol}-USDT_SMA'] = sma(dfs[f'{symbol}-USDT_Open'],MA_window)
    dfs[f'{symbol}-USDT_UpperBand'],dfs[f'{symbol}-USDT_LowerBand'] = bollinger_bands(dfs[f'{symbol}-USDT_Open']
                                                                                              ,dfs[f'{symbol}-USDT_SMA']
                                                                                              ,MA_window
                                                                                              ,nsdt)
dfs.dropna(inplace=True)
dfs
```

```python
class TradingEnv:
    def __init__(self, balance_amount, balance_unit, trading_fee_multiplier):
        self.balance_amount = balance_amount
        self.balance_unit = balance_unit
        self.buys = []
        self.sells = []
        self.paid_fees = []
        self.trading_fee_multiplier = trading_fee_multiplier #VIP level 0 paying fees with BNB
        
        
    def buy(self,symbol,buy_price,time):
        self.paid_fees.append(self.balance_amount * (1 - self.trading_fee_multiplier))
        self.balance_amount = (self.balance_amount / buy_price) * self.trading_fee_multiplier
        self.balance_unit = symbol
        self.buys.append([symbol,time,buy_price])
        
            
    def sell(self,sell_price,time):
        self.paid_fees.append((self.balance_amount * sell_price) * (1 - self.trading_fee_multiplier))    
        self.balance_amount = self.balance_amount * sell_price * self.trading_fee_multiplier
        self.sells.append([self.balance_unit,time,sell_price])
        self.balance_unit = 'USDT'
```

```python
initial_balance = 100
info = []
final_balance = []
buys = {}
sells = {}

################## BOLLINGER BANDS STRATEGY ######################
for symbol in symbols:
    env = TradingEnv(balance_amount=100,balance_unit='USDT',trading_fee_multiplier = 0.99925)
    
    for i in range(len(dfs)):
        if env.balance_unit == 'USDT': 
            if dfs[f'{symbol}-USDT_Low'].iloc[i] < dfs[f'{symbol}-USDT_LowerBand'].iloc[i]: 
                env.buy(symbol, dfs[f'{symbol}-USDT_LowerBand'].iloc[i], dfs['OpenTime'].iloc[i])

        if (env.balance_unit != 'USDT'):  
            if (dfs[f'{symbol}-USDT_High'].iloc[i]) > (dfs[f'{symbol}-USDT_UpperBand'].iloc[i]):
                env.sell((dfs[f'{symbol}-USDT_UpperBand'].iloc[i]), (dfs['OpenTime'].iloc[i]))

    if env.balance_unit != 'USDT':
        env.sell(dfs[f'{symbol}-USDT_Close'].iloc[-1],dfs['CloseTime'].iloc[-1])

################ RESULT DATA ######################

    results = [symbol,len(env.buys),len(env.sells),np.sum(env.paid_fees),initial_balance,env.balance_amount,
               (env.balance_amount-initial_balance),(((env.balance_amount/initial_balance)-1)*100)]
    info.append(results)
    
    total_balance = initial_balance * len(symbols)
    final_balance.append(env.balance_amount)
    
    buys[symbol] = env.buys
    sells[symbol] = env.sells

frame = pd.DataFrame(info)
frame.columns = ['Coin','# Buys','# Sells','Fees $','Intial Balance $','Final Balance $','Profit $','Profit %']
frame
```
